```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import csv
```

## Define

```python
#TRAIN_DATA_FILES = ['cross','dead_end', 'left', 'right', 'straight', 'threeway_left'. 'threeway_center', 'threeway_right']
TRAIN_DATA_FILE = "threeway_center"
NUM_CLASSES = len(TRAIN_DATA_FILE)
MAX_RANGE = 180 # [deg]

REPLACE_NAN = 0.0

RESOLUTION = 20.0 #(cm)
WIDTH = HEIGHT = int(3000 * 2 / RESOLUTION + 2) # 3000 = 30[m]
```

## 学習データ読み込み用関数

```python
def split_data():
#    files = os.listdir('/content/drive/My Drive/Colab Notebooks/data')
    X = []

    for i in range(NUM_CLASSES):
        try:
          # header = 列名
            file_name = "/content/drive/My Drive/Colab Notebooks/data2/" + TRAIN_DATA_FILE + ".csv"
            data_set = pd.read_csv(file_name, header=None)
        except pd.io.common.EmptyDataError:
            print("ERROR: {} is empty".format(file_name))

#    X = pd.concat(data_set)
    # replace Nan with 'REPLACE_NAN'
    X = data_set.fillna(REPLACE_NAN)

    return data_set, X.values
```

## Plot

```python
read_file = open('/content/drive/My Drive/Colab Notebooks/data2/' + TRAIN_DATA_FILE + '.csv', 'a')
output_file = open('/content/drive/My Drive/Colab Notebooks/data2/new_' + TRAIN_DATA_FILE + '.csv', 'a')

dir_output_file = '/content/drive/My Drive/Colab Notebooks/data2/new_' + TRAIN_DATA_FILE + '.csv'
```

```python
X, X_val = split_data()
X_val *= 100 / RESOLUTION
_, DIM_input_data = X_val.shape

rotate = 0

matrix_x = [ [-np.sin(rotate + np.deg2rad((-MAX_RANGE/2) + (i / DIM_input_data * MAX_RANGE))) for i in range(DIM_input_data)]  ]
matrix_y = [ [np.cos(rotate + np.deg2rad((-MAX_RANGE/2) + (j / DIM_input_data * MAX_RANGE))) for j in range(DIM_input_data)]  ]

point_cloud_x = np.round(X_val * matrix_x) + (WIDTH / 2)
point_cloud_y = np.round(X_val * matrix_y) + (HEIGHT / 2)
```

```python
#counter==0 のとき1回目のラベリングは必ずスキップしないとエラー
num_skip = 9
batch = 15000
for i in range(counter*batch+tmp, (counter+1)*batch, num_skip):
  fig, ax = plt.subplots(figsize=(10, 10))
  plt.scatter(point_cloud_x[i], point_cloud_y[i], color="k")
  plt.grid(True)
  ax.set_xlim([0, WIDTH])
  ax.set_ylim([0, HEIGHT])

#  plt.axis("off")
  plt.show()
  flg = 0
  print(str(i) + ' times')
  flg = input()
  if(flg == '0'):
      print('remove')
  elif(flg == 'q'):
    tmp = i - counter*batch
    print('quit')
    break
  else:
    for j in range(num_skip):
      pd.DataFrame(X.iloc[i-j]).T.to_csv(dir_output_file, mode='a', header=False, index=False)
    print('skip')

if(i >= (counter+1)*batch-num_skip):
  tmp = 0
  counter += 1
print("counter =", counter)
```
