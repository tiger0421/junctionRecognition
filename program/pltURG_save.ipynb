```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import csv
```

## Define

```python
#TRAIN_DATA_FILES = ['cross','dead', 'left', 'right', 'straight', 'threeway']
TRAIN_DATA_FILES = ['new_4-4']
OUTPUT_DATA_FILES = [ 'straight', 'dead_end', 'right','left', 'cross', 'threeway_right', 'threeway_center', 'threeway_left']

# NUM_CLASSES = 6
NUM_CLASSES = len(TRAIN_DATA_FILES)
MAX_RANGE = 180 # [deg]

REPLACE_NAN = 0.0

RESOLUTION = 20.0 #(cm)
WIDTH = HEIGHT = int(3000 * 2 / RESOLUTION + 2)  # 3000 = 30[m]
```

```python
def adjust_data_num(num_class):
    # header = 列名
    file_name = "/content/drive/My Drive/Colab Notebooks/data2/" + TRAIN_DATA_FILES[num_class] + ".csv"
    data_set = pd.read_csv(file_name, header=None)

    return data_set
```

## 学習データ読み込み用関数

```python
def split_data():
    X = []
    all_data_set = []

    for i in range(NUM_CLASSES):
        try:
            data_set = adjust_data_num(i)
            all_data_set.append(data_set)

        except pd.io.common.EmptyDataError:
            print("ERROR: {} is empty".format(file_name))

    X = pd.concat(all_data_set)
    # replace Nan with 'REPLACE_NAN'
    X = X.fillna(REPLACE_NAN)

#    _, DIM_input_data = data_set.shape

    return data_set, X.values
```

## Plot

```python
X, X_val = split_data()
X_val *= 100 / RESOLUTION
_, DIM_input_data = X_val.shape

rotate = 0

matrix_x = [ [-np.sin(rotate + np.deg2rad((-MAX_RANGE/2) + (i / DIM_input_data * MAX_RANGE))) for i in range(DIM_input_data)]  ]
matrix_y = [ [np.cos(rotate + np.deg2rad((-MAX_RANGE/2) + (j / DIM_input_data * MAX_RANGE))) for j in range(DIM_input_data)]  ]

point_cloud_x = np.round(X_val * matrix_x) + (WIDTH / 2)
point_cloud_y = np.round(X_val * matrix_y) + (HEIGHT / 2)

```
